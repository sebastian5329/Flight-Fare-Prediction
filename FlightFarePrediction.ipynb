```python
data1["Total_Stops"]=data1["Total_Stops"].replace(1,"1 stop")
data1["Total_Stops"].value_counts()
```

```python
data1["Total_Stops"]=data1["Total_Stops"].replace({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4})
data1
```

```python
#converting Date_of_Journey to datetime
data1["Date_of_Journey"]=pd.to_datetime(data1["Date_of_Journey"],dayfirst=True)
```

```python
data1["Journey_Month"]=data1["Date_of_Journey"].dt.month          #Finding the month of travel
```

```python
data1["Day_of_Week"]=data1["Date_of_Journey"].dt.dayofweek
```

```python
data1["Journey_Year"]=data1["Date_of_Journey"].dt.year
```

```python
data1
```

```python
#converting duration into minutes
data1["Duration"]=data1["Duration"].astype(str) 
def dur_min(Duration):
    Duration=Duration.lower()
    parts=Duration.split()
    total=0
    for part in parts:
        if 'h'in part:
            total+= int(part.replace("h",""))*60
        elif 'm' in part:
            total+= int(part.replace("m",""))
    return total
data1['Duration_Minutes']=data1["Duration"].apply(dur_min)
data1
```

```python
data1["Arrival_Time"]=data1["Arrival_Time"].apply(lambda x: x.split()[0])
```

```python
data1["Dep_Time"]=data1["Dep_Time"].apply(lambda x: x.split()[0]) #lambda fn is the easiest way to apply a mini_fn in pandas dataframe column(here for each value of x ,x is split and only the first term is taken.
```

```python
data1
```

```python
data1["Route"]=data1["Route"].fillna("None")
route_splits=data1["Route"].str.split('→',expand=True)
route_splits.columns=[f'Route_{i+1}'for i in range(route_splits.shape[1])]
data1=pd.concat([data1,route_splits],axis=1)
```

```python
l1=['Route','Duration']
data1.drop(l1,axis=1,inplace=True)
data1
```

```python
data1
```

```python
data1.drop('Date_of_Journey',axis=1)
```

```python
data1["Dep_Time"]=pd.to_datetime(data1["Dep_Time"],format='%H:%M')
data1["Arrival_Time"]=pd.to_datetime(data1["Arrival_Time"],format='%H:%M')
data1["Dep_hour"]=data1["Dep_Time"].dt.hour
data1["Dep_min"]=data1["Dep_Time"].dt.minute 
data1["Arr_hour"]=data1["Arrival_Time"].dt.hour
data1["Arr_min"]=data1["Arrival_Time"].dt.minute
data1
```

```python
from sklearn.preprocessing import StandardScaler,FunctionTransformer,OrdinalEncoder,LabelEncoder
```

```python

standard_scaler=['Duration_Minutes']
route_cols=['Route_1','Route_2','Route_3','Route_4','Route_5','Route_6']
Total_Stops_enc=OrdinalEncoder(categories=[[0,1,2,3,4]])
Pass_through=['Addinfo_No info','Addinfo_In-flight meal not included','Addinfo_No check-in baggage included','Addinfo_1 Short layover','Addinfo_Change airports','Addinfo_Business class','Addinfo_Red-eye flight','Addinfo_2 Long layover','Journey_Month','Day_of_Week','Dep_hour','Dep_min','Arr_hour','Arr_min','Airline_Air India','Airline_GoAir','Airline_IndiGo','Airline_Jet Airways','Airline_Jet Airways Business','Airline_Multiple carriers','Airline_Multiple carriers Premium economy','Airline_SpiceJet','Airline_Trujet','Airline_Vistara','Airline_Vistara Premium economy','Source_Chennai','Source_Delhi','Source_Kolkata','Source_Mumbai','Destination_Cochin','Destination_Delhi','Destination_Hyderabad','Destination_Kolkata']
```

```python
def same(x):
    return x
```

```python
no_trans=FunctionTransformer(same)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor=ColumnTransformer([("Passthrough",no_trans,Pass_through),("standard_scaler",StandardScaler(),standard_scaler),("Ordinal_Routes",OrdinalEncoder(),route_cols),("Ordinal_Total_Stops",Total_Stops_enc,["Total_Stops"])])
```
